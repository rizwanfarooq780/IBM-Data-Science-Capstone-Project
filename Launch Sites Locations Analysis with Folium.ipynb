```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium import DivIcon
from math import radians, cos, sin, sqrt, atan2
import pandas as pd

# Define the Haversine formula for distance calculation
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in kilometers
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Coordinates for Site A (the launch site)
site_a = {'Description': 'Launch Site', 'Latitude': 28.56342, 'Longitude': -80.57674}

# Data for other locations (Highway, Florida City, Coastline)
data = {
    'Description': ['Highway', 'Florida City', 'Coastline'],
    'Latitude': [28.411780, 28.5383, 28.56342],
    'Longitude': [-80.820630, -81.3792, -80.56756]
}

# Create a DataFrame for the locations
locations_df = pd.DataFrame(data)

# Initialize the map centered around the launch site
site_map = folium.Map(location=[site_a['Latitude'], site_a['Longitude']], zoom_start=9)

# Marker for Launch Site
folium.Marker(
    [site_a['Latitude'], site_a['Longitude']],
    popup='Launch Site',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(site_map)

# Add markers and lines from Launch Site to each location
for index, row in locations_df.iterrows():
    # Marker for each location
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup=row['Description'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(site_map)

    # Line from Launch Site to the location
    line = folium.PolyLine(locations=[[site_a['Latitude'], site_a['Longitude']], [row['Latitude'], row['Longitude']]], color='blue', weight=2)
    site_map.add_child(line)

    # Calculate and display distance
    distance = calculate_distance(site_a['Latitude'], site_a['Longitude'], row['Latitude'], row['Longitude'])
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        icon=DivIcon(
            icon_size=(200,36),
            icon_anchor=(0,0),
            html='<div style="font-size: 12px; color: blue;">{:.2f} km</div>'.format(distance)
        )
    ).add_to(site_map)

# Display the map
site_map.save('launch_site_to_locations_map.html')
site_map
```
